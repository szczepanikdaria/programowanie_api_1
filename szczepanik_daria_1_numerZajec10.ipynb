```python
import spacy
```

```python
nlp = spacy.load("pl_core_news_sm")
```

```python
text = "Gdy Maria zobaczyła smoka, postanowiła wrócić do zamku w Warszawie."
```

```python
doc = nlp(text)
```

```python
type(doc)
```

```python
nlp.pipeline
```

```python
for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.shape_, token.is_stop)
```

```python
#NER - Named Entity Recognition
for ent in doc.ents:
    print(ent.text, ent.label_)
```
